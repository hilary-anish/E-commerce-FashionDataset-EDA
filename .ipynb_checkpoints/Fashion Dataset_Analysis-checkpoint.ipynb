### 1. Data Cleaning process

##### -- The first column doesn't have any significance, hence can be removed

```python
df.drop(columns=['Unnamed: 0'], inplace = True)
```

##### -- Replace 'Size:', 'Rs\n' and '% off' in the columns 'Sizes' , 'MRP' and 'Discount' respectively

```python
df['Sizes'] = df['Sizes'].str.replace('Size:',' ')
```

```python
df['MRP'] = df['MRP'].str.replace('Rs\n',' ')
```

```python
df['Discount'] = df['Discount'].str.replace('% off',' ')
```

##### -- Correct the spelling of Details column and change other column names

```python
df.rename(columns = {'Deatils':'Details', 'Discount':'Discount in %', 'MRP':'MRP in Rs','SellPrice':'SellPrice in Rs'}, inplace = True)
```

```python
df.head()
```

#### Handling null values

```python
df.replace(to_replace = 'Nan', value = np.nan, inplace=True)
```

```python
df.isnull().sum()
```

##### -- Ploting the null values

```python
sns.heatmap(df.isnull(), vmin = 0,vmax = 1, yticklabels = False)
```

##### -- Droping the Nan values from BrandName

```python
df.drop(df.loc[df['BrandName'].isnull()].index, inplace = True)
```

```python
sns.heatmap(df.isnull(), vmin = 0,vmax = 1, yticklabels = False)
```

```python
df.isnull().sum()
```

##### -- Handling the MRP Nan values by using the mean MRP price of each Brand

```python
df.info()
```

##### -- Converting object types to float64

```python
df['MRP in Rs'] = df['MRP in Rs'].astype(np.float64)
df['SellPrice in Rs'] = df['SellPrice in Rs'].astype(np.float64)
df['Discount in %'] = df['Discount in %'].astype(np.float64)
```

```python
df.info()
```

```python
Brand_list = list(df.BrandName.value_counts().index)
```

```python
# Storing the min and max MRP value of each Brand in a dictionary

MRP = {}
count = 0
for brand in Brand_list:
    if count< 10:
        temp = df.loc[df.BrandName == brand ,'MRP in Rs']
        count+=1
        MRP.update({brand:[int(temp.min()),int(temp.max())]})
```

```python
MRP
```

```python
import itertools

x_keys = []
y_values = []

for k,v in MRP.items():
    x_keys.extend(list(itertools.repeat(k,len(v))))
    y_values.extend(v)

plt.figure(figsize = (13,5))
sns.barplot(x = x_keys,y = y_values)
```

#### The barchart shows the price range of top 10 Brands with most collection

```python
# Getting the mean MRP value of each Brand to fill the null values

MRP_mean = {}
for brand in Brand_list:
        temp = df.loc[df.BrandName == brand ,'MRP in Rs']
        MRP_mean.update({brand:temp.mean()})
```

##### -- Applying the mean MRP price of each Brand for the corresponding Nan values

```python
def is_missing(MRP,Brand,mean):    
    if np.isnan(MRP):
        return mean[Brand]
    else:
        return MRP

df['MRP in Rs'] = df.apply(lambda x: is_missing(x['MRP in Rs'],x['BrandName'],MRP_mean),axis=1)
```

```python
df.isnull().sum()
```

#### The above value shows, the Nan values reduced from 7025 but there are 2188 Brands without MRP values but only Selling price

##### -- Top brands for each of the most sold Categories

```python
df_cat = df.groupby(['Category','BrandName']).size().reset_index().rename(columns = {0:'Items'})
cat_list = list(df['Category'].value_counts().index)
print(cat_list)
```

```python
# Dictionary of different categories with their top brands

brand_cat = {}
val_brand = []
for cat in cat_list:
    itms = df_cat.loc[df_cat['Category'] == cat,'Items']
    idx = itms.nlargest(10).index.tolist()
    val = itms.nlargest(10).tolist()
    brand = df_cat.loc[idx,'BrandName']
    val_brand.append(val)
    brand_cat.update({cat:brand.tolist()})
```

##### -- Visulaizing categories with their top brands

```python
brand_x = []
cat_name = []
for k,v in brand_cat.items():
    brand_x.append(v)
    cat_name.append(k)
    
c = 0
for i in range(len(brand_x)):
    plt.figure(c, figsize = (6,6))
    plt.pie(val_brand[i], labels = brand_x[i])
    plt.title(cat_name[i])
    c+=1
```

### Popular brands for each of the categories have been captured in the above pie charts

##### --The function returns the brand name along with the cheapest and the highest price in the Category

```python
def cheap_costly(num,cat_type):
    df_sell = df.groupby(['Category','SellPrice in Rs','BrandName']).size().reset_index()
    min_price = df_sell.loc[df_sell['Category'] == cat_type, 'SellPrice in Rs'].nsmallest(num)
    min_idx = min_price.index.tolist()
    max_price = df_sell.loc[df_sell['Category'] == cat_type, 'SellPrice in Rs'].nlargest(num)
    max_idx = max_price.index.tolist()

    brand_min = df_sell.loc[min_idx,'BrandName'].tolist()
    brand_max = df_sell.loc[max_idx,'BrandName'].tolist()
    maxp = max_price.tolist()
    minp = min_price.tolist()

    mini = dict(zip(minp,brand_min))
    maxi = dict(zip(maxp,brand_max))
    
    return (mini,maxi)
```

```python
cheap_costly(2,'Footwear-Women')
```

### The above function helps the customer to identify the brand based on the budget or vice versa
